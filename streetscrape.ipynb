```python
# Find orgs
fout = open('foundcats.csv', 'w')
for i in range(1, ws.max_row):
    cell = ws['A'+str(i)]
    if cell.font.sz == 36:
        print('{} CATEGORY: {}'.format(i, cell.value))
        fout.write('\n{} CATEGORY: {}\n'.format(i, cell.value))
    if cell.font.sz == 18:
        print('{} SUBCATEGORY: {}'.format(i, cell.value))
        fout.write('{} SUBCATEGORY: {}\n'.format(i, cell.value))
    if str(cell.value).isupper():
        if cell.value.find('–') != 0:
            print('{} ORG: {}'.format(i, cell.value))
            fout.write('{} ORG: {}\n'.format(i, cell.value))
fout.close()
```

```python
# Scrape dataset

def get_org_description(ws, orgstart, orgend):
    #description = [' '.join([ws['A'+str(j)].value for j in [orgstart+1,orgend]])]
    description = ''
    for j in [orgstart+1,orgend]:
        cell = ws['A'+str(j)]
        if cell.data_type != 'n':
            description += ' ' + cell.value
    print('description: {}'.format(description))
    return description

orgstart = -1
orgend = -1
category = ''
subcategory = ''
organisation = ''
rows = []
for i in range(1, ws.max_row):
    cell = ws['A'+str(i)]
    if cell.font.sz == 36:
        if orgstart > -1:
            rows += [[category, subcategory, organisation, 
                      get_org_description(ws, orgstart, i-1)]]
        category = cell.value
        print('{} CATEGORY: {}'.format(i, category))
        continue

    if cell.font.sz == 18:
        if orgstart > -1:
            rows += [[category, subcategory, organisation, 
                      get_org_description(ws, orgstart, i-1)]]
        subcategory = cell.value
        print('{} SUBCATEGORY: {}'.format(i, subcategory))
        continue
    
    if str(cell.value).isupper():
        if cell.value.find('–') != 0:
            if orgstart > -1:
                print('from {} to {}'.format(orgstart, i-1))
                rows += [[category, subcategory, organisation, 
                          get_org_description(ws, orgstart, i-1)]]
            organisation = cell.value
            print('{} ORG: {}'.format(i, organisation))
            orgstart = i+1

df = pd.DataFrame(rows, columns=['category', 'subcategory', 'name', 'description'])
df
```
